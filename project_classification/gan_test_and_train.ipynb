** Step 2: Link Images with Labels and Preprocess Image Data**

```python
import os
import cv2
import json

# Paths
json_path = r"C:\MurthyLab\project_classification\classification.json"
image_folder = r"C:\MurthyLab\project_classification\Water melon"

# Load JSON labels
def load_labels(json_path):
    with open(json_path, "r") as file:
        data = json.load(file)  # Parse JSON
    
    labels_dict = {}
    for entry in data:
        image_name = entry.get("file_upload", "").lower()  
        annotations = entry.get("annotations", [])

        if annotations and "result" in annotations[0]:
            label_choices = annotations[0]["result"][0]["value"].get("choices", [])
            if label_choices:
                labels_dict[image_name] = label_choices[0]  

    return labels_dict  

labels_dict = load_labels(json_path)

# Normalize filenames in image folder
def clean_filename(filename):
    parts = filename.split("-")
    return parts[-1].lower()

image_filenames = {clean_filename(img): os.path.join(image_folder, img) for img in os.listdir(image_folder)}

# Link images to labels
image_data = []
for json_filename, label in labels_dict.items():
    cleaned_name = clean_filename(json_filename)  
    if cleaned_name in image_filenames:
        image_data.append((image_filenames[cleaned_name], label))  

print(f"\n Linked {len(image_data)} images with labels!")

# Preprocess images
def preprocess_image(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    if image is None:
        print(f" Error: Image not found - {image_path}")
        return None
    
    image = cv2.resize(image, (256, 256))
    image = image / 255.0
    
    return image

# Apply preprocessing
X = [preprocess_image(path) for path, _ in image_data if preprocess_image(path) is not None]

print(f"\n Successfully preprocessed {len(X)} images!")
```

** Step 4: Define and Train a UNet Autoencoder Model**

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
import cv2
import numpy as np

# Define UNet Architecture
class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()
        
        # Encoder
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)

        # Decoder
        self.upconv1 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.upconv2 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.conv_out = nn.Conv2d(64, 3, kernel_size=3, padding=1)

        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # Encoder
        x1 = self.relu(self.conv1(x))
        x2 = self.pool(self.relu(self.conv2(x1)))
        x3 = self.pool(self.relu(self.conv3(x2)))

        # Decoder
        x3 = self.upconv1(x3)
        x2 = x3 + x2  # Skip connection
        x2 = self.upconv2(x2)
        x1 = x2 + x1  # Skip connection
        x_out = self.sigmoid(self.conv_out(x1))

        return x_out

# Load Dataset
class WatermelonDataset(Dataset):
    def __init__(self, images, labels, transform=None):
        self.images = images
        self.labels = labels
        self.transform = transform
    
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        image = self.images[idx]
        label = self.labels[idx]
        
        if self.transform:
            image = self.transform(image)
        
        return image, label

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

dataset = WatermelonDataset(X, [label for _, label in image_data], transform=transform)
dataloader = DataLoader(dataset, batch_size=8, shuffle=True)

# Initialize Model, Loss, Optimizer
model = UNet()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train UNet Model
epochs = 20  # Increase epochs for better learning
for epoch in range(epochs):
    for images, _ in dataloader:
        images = images.float()
        optimizer.zero_grad()
        reconstructed = model(images)
        loss = criterion(reconstructed, images)
        loss.backward()
        optimizer.step()
    
    print(f" Epoch {epoch+1}/{epochs} - Loss: {loss.item():.4f}")

print(" Training Complete!")
```

**Simple UNet Autoencoder for Image Reconstruction  Training on Dummy Data and Inference on a Test Image**

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
import cv2
import numpy as np

# --- Step 1: Define UNet Model ---
class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()
        
        # Encoder
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)

        # Decoder
        self.upconv1 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.upconv2 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.conv_out = nn.Conv2d(64, 3, kernel_size=3, padding=1)

        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # Encoder
        x1 = self.relu(self.conv1(x))
        x2 = self.pool(self.relu(self.conv2(x1)))
        x3 = self.pool(self.relu(self.conv3(x2)))

        # Decoder
        x3 = self.upconv1(x3)
        x2 = x3 + x2  # Skip connection
        x2 = self.upconv2(x2)
        x1 = x2 + x1  # Skip connection
        x_out = self.sigmoid(self.conv_out(x1))

        return x_out

# --- Step 2: Prepare Dataset (dummy example with random data) ---
class DummyDataset(Dataset):
    def __init__(self, size=100):
        self.size = size
        self.transform = transforms.Compose([
            transforms.ToTensor(),
        ])
        
    def __len__(self):
        return self.size
    
    def __getitem__(self, idx):
        # Random image tensor (3x256x256)
        image = np.random.rand(256, 256, 3).astype(np.float32)
        image = self.transform(image)
        label = image  # Autoencoder target = input image
        return image, label

dataset = DummyDataset()
dataloader = DataLoader(dataset, batch_size=8, shuffle=True)

# --- Step 3: Initialize model, loss, optimizer ---
model = UNet()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# --- Step 4: Train model (quick training loop) ---
epochs = 2  # Use more epochs for real training

for epoch in range(epochs):
    model.train()
    total_loss = 0
    for images, labels in dataloader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss/len(dataloader):.4f}")

# --- Step 5: Save model weights ---
save_path = "unet_model_weights.pth"
torch.save(model.state_dict(), save_path)
print(f"Model saved to {save_path}")

# --- Step 6: Load model weights for evaluation ---
loaded_model = UNet()
loaded_model.load_state_dict(torch.load(save_path))
loaded_model.eval()

# --- Step 7: Evaluate on a new image (dummy example) ---
def preprocess_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.resize(image, (256, 256))
    image = image / 255.0
    image = torch.tensor(image, dtype=torch.float32).permute(2, 0, 1).unsqueeze(0)  # (1,3,256,256)
    return image

# Use your own test image path here:
test_image_path = r"C:\MurthyLab\project_classification\Unseendata\3_jpg.rf.aa73f968045f2a2468de219ec2b78a34.jpg"

if os.path.exists(test_image_path):
    input_tensor = preprocess_image(test_image_path)

    with torch.no_grad():
        output_tensor = loaded_model(input_tensor)

    print("Inference done! Output shape:", output_tensor.shape)
else:
    print(f"Test image not found at: {test_image_path}")
```
